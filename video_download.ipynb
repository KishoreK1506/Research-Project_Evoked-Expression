```python
import os
import pytube
from pytube.exceptions import VideoUnavailable
from tqdm import tqdm
import pandas as pd
```

```python
train_folder = "/Users/kj/Desktop/Research Project/Train videos"
test_folder = "/Users/kj/Desktop/Research Project/Test Video"
```

```python
data_train = pd.read_csv("/Users/kj/Desktop/Research Project/Data set/train.csv")
data_test = pd.read_csv("/Users/kj/Desktop/Research Project/Data set/test.csv")
```

```python
video_ids_train = data_train["YouTube ID"].unique()
video_ids_test = data_test["Video ID"].unique()
```

```python
if not os.path.exists(train_folder):
    os.mkdir(train_folder)
if not os.path.exists(test_folder):
    os.mkdir(test_folder)
```

```python
for id_ in tqdm(video_ids_train):
    print(id_, "  :: DOWNLOADING")
    try:
        output_path = train_folder + "/" + "{}".format(id_)
        if os.path.exists(output_path):
            continue
            
        youtube = pytube.YouTube("https://www.youtube.com/watch?v=" + id_)
        streams = youtube.streams.filter(progressive=True, file_extension='mp4')
        video = streams.get_by_resolution("360p")
        # Download the video
        video.download(output_path)
        print(id_, "  :: DOWNLOADED")
    except Exception as e : 
        print(id_, "  :: ERROR DOWNLOADING" ,e)
    
```
